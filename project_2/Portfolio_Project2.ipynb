## So we chack the data and now lets clean it

1- we will delete the 'index' column cuz we dont need it

2- change the data type of 'Base Pay' to flout , and delete the symbol '$'

```python
#Delete the index column
data_df.drop(columns=['index'] , inplace=True)
```

```python
data_df['Base Pay'].value_counts()      #we need to delete $
```

```python
data_df['Base Pay'] = data_df['Base Pay'].str.replace('$', '', regex=False)     #replaece the remove $ and set the '' 
```

```python
data_df['Base Pay'].dtypes
```

```python
data_df['Base Pay'] = data_df['Base Pay'].astype('float')       #change the datatype to flout
```
