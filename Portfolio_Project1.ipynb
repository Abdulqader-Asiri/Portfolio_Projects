lets add the Profit and ROMS

profit Formula is (Profit = Revenue - Marketing Spend)

```python
databasse['profit'] = databasse['Revenue'] - databasse["Marketing Spend"]
databasse
```

```python
databasse.sample(4)
```

ROMS Formula is (ROMS = (profit / Marketing Spend)* 100)

```python
databasse['ROMS'] = round((databasse['profit'] / databasse["Marketing Spend"]) * 100, 2)
```

```python
databasse['ROMS'].round(2)
```

ROMS% Formula is (ROMS% = (ROMS%) / 100 , 2)

```python
databasse['ROMS%'] = round(databasse['ROMS'] / 100 , 2)
databasse.head(5)
```
