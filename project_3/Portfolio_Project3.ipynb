#### lets create a new columns for year,month,week,day

```python
coved_df['year'] = ss.DatetimeIndex(coved_df.date).year
coved_df['month'] = ss.DatetimeIndex(coved_df.date).month
coved_df['day'] = ss.DatetimeIndex(coved_df.date).day
coved_df['weekday'] = ss.DatetimeIndex(coved_df.date).weekday
coved_df        #add some data frame for months , year, day,week
```

```python
coved_df['total_cases'] = coved_df.new_cases.cumsum()
coved_df['total_deaths'] = coved_df.new_deaths.cumsum()
coved_df['total_tests'] = coved_df.new_tests.cumsum() + before_dily
coved_df
```

## lets add ITALY(locations) to the data

```python
urlretrieve('https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv', 
            'c:\\Users\\file Path\\locations.csv')        #data for locations
```

```python
locations_df = ss.read_csv('c:\\Users\\file Path\\locations.csv')
locations_df
```

```python
locations_df[locations_df.location == "Italy"]
```

```python
coved_df['location'] = "Italy"      #add new column and name it ITALY
coved_df
```

```python
merget_df = coved_df.merge(locations_df, on='location')     #merge the tow data
merget_df
```

```python
merget_df['cases_per_million'] = merget_df.total_cases * 1e6 / merget_df.population
merget_df['deaths_per_million'] = merget_df.total_deaths * 1e6 / merget_df.population
merget_df['tests_per_million'] = merget_df.total_tests * 1e6 / merget_df.population
merget_df
```
